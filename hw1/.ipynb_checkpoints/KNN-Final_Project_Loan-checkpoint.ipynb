## k-NN Algorithm with Normalization


```python
import numpy as np
from collections import Counter

def calculate_accuracy(y_true, y_pred):
    correct = sum(1 for true, pred in zip(y_true, y_pred) if true == pred)
    return correct / len(y_true)

def calculate_f1_score(y_true, y_pred):
    true_positive = sum(1 for true, pred in zip(y_true, y_pred) if true == pred == 1)
    predicted_positive = sum(1 for pred in y_pred if pred == 1)
    actual_positive = sum(1 for true in y_true if true == 1)
    precision = true_positive / predicted_positive if predicted_positive > 0 else 0
    recall = true_positive / actual_positive if actual_positive > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0
    return f1
def features_normalization(X):
    X_num = X.apply(pd.to_numeric) #Typecasting to Numeric
    min_value = X_num.min()
    max_value = X_num.max()
    X_normalized = (X_num - min_value) / (max_value - min_value)
    return X_normalized

def distance(X1, X2):
    Euclidean_distance = np.sqrt(np.sum((X1 - X2) ** 2)) # Euclidean distance calculation
    return Euclidean_distance

def knn_predict(X_train, y_train, X_test, k):
    final_output = []
    for i in range(len(X_test)):
        distances = [(distance(X_train.iloc[j], X_test.iloc[i]), j) for j in range(len(X_train))]
        distances.sort()
        neighbors = distances[:k]
        neighbor_labels = [y_train.iloc[j[1]] for j in neighbors]
        predicted_label = Counter(neighbor_labels).most_common(1)[0][0]
        final_output.append(predicted_label)
    return final_output
```

```python
# Define your k value and other parameters
k_values = list(range(1, 51, 2))
iterations = 20
k_fold = 10

for k in k_values:
    accuracy_values_train = []
    accuracy_values_test = []
    f1_scores_train = []
    f1_scores_test = []

    for i in range(iterations):
        # Shuffle data
        df_parkinsons_shuffled = shuffle(df_parkinsons)

        # Split data into k folds
        fold_size = len(df_parkinsons_shuffled) // k_fold
        for fold in range(k_fold):
            test_start = fold * fold_size
            test_end = (fold + 1) * fold_size
            test_df = df_parkinsons_shuffled.iloc[test_start:test_end]
            train_df = pd.concat([df_parkinsons_shuffled.iloc[:test_start], df_parkinsons_shuffled.iloc[test_end:]])

            X_train = train_df.drop(columns=[target_column])
            y_train = train_df[target_column]
            X_test = test_df.drop(columns=[target_column])
            y_test = test_df[target_column]
            X_train_normalized = features_normalization(X_train)
            X_test_normalized = features_normalization(X_test)

            # Training the k-NN algorithm
            predictions_train = knn_predict(X_train_normalized, y_train, X_train_normalized, k)
            accuracy_train = calculate_accuracy(y_train, predictions_train)
            f1_train = calculate_f1_score(y_train, predictions_train)
            accuracy_values_train.append(accuracy_train)
            f1_scores_train.append(f1_train)
            #print("accuracy_values_train: ",accuracy_values_train)
            
            # Evaluating the k-NN model
            predictions_test = knn_predict(X_train_normalized, y_train, X_test_normalized, k)
            accuracy_test = calculate_accuracy(y_test, predictions_test)
            f1_test = calculate_f1_score(y_test, predictions_test)
            accuracy_values_test.append(accuracy_test)
            f1_scores_test.append(f1_test)

    # Compute average performance metrics for training set
    avg_accuracy_train = np.mean(accuracy_values_train)
    avg_f1_train = np.mean(f1_scores_train)

    # Compute average performance metrics for testing set
    avg_accuracy_test = np.mean(accuracy_values_test)
    avg_f1_test = np.mean(f1_scores_test)
```
